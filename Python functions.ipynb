```python
def calculate_area(length, width):
    area = length * width
    return area
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
area = calculate_area(length, width)
print(f"The area of the rectangle is: {area}")
```

```python
def power(base, exponent):
    return base ** exponent
base = float(input("Enter the base number: "))
exponent = float(input("Enter the exponent: "))
result = power(base, exponent)
print(f"{base} raised to the power of {exponent} is: {result}")
```

```python
def get_full_name(first, middle, last):
    return f"{first} {middle} {last}".strip()
first_name = str(input("Enter your first name: "))
middle_name = str(input("Enter your middle name: "))
last_name = str(input("Enter your last name: "))
full_name = get_full_name(first_name, middle_name, last_name)
print(f"Your full name is: {full_name}")
```

```python
def celsius_to_farenheit(celsius):
    return celsius * 9/5 + 32
def farenheit_to_celsius(farenheit):
    return (farenheit - 32) * 5/9
temp_c = float(input("Enter temperature in Celsius: "))
temp_f = celsius_to_farenheit(temp_c)
print(f"{temp_c}°C is equal to {temp_f}°F")
temp_f = float(input("Enter temperature in Farenheit: "))
temp_c = farenheit_to_celsius(temp_f)
print(f"{temp_f}°F is equal to {temp_c}°C")
```

```python
def analyze_temperature(value, unit):
    if unit.upper() == "F":
        value = fahrenheit_to_celsius(value)
    if value < 0:
        return "Freezing" 
    elif value <= 15:
        return "Cold"
    elif value <= 25:
        return "Moderate"
    else:
        return "Hot"
temp_value = float(input("Enter the temperature value: "))
temp_unit = str(input("Enter the temperature unit (C/F): "))
result = analyze_temperature(temp_value, temp_unit)
print(f"The temperature is considered: {result}")
```

```python
def calculate_average(numbers):
    return sum(numbers) / len(numbers)
numbers = [10,20,30,40,50]
average = calculate_average(numbers)
print(f" The average of the numbers is: {average}")
```

```python
def calculate_final_grade(student_name, scores):
    if not scores:
        return f"{student_name} has no scores available."
    avg_score = sum(scores) / len(scores)
    if avg_score >= 90:
        grade = "A"
    elif avg_score >= 80:
        grade = "B"
    elif avg_score >= 70:
        grade = "C"
    elif avg_score >= 60:
        grade = "D"
    else:
        grade = "F"
    return f"{student_name}'s final grade is: {avg_score:.1f} ({grade})"
student_name = str(input("Enter the student name: "))
scores_input = str(input("Enter the student's scores separated by commas: "))
scores = [float(x.strip()) for x in scores_input.split(",")]
print(scores)
result = calculate_final_grade(student_name, scores)
print(result)

```
