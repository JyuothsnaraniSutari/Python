```python
df['email']=df['email'].fillna(df['email'].mode())
df['phone']=df['phone'].fillna(df['phone'].mode())
df['zip_code']=df['zip_code'].ffill()
df['registration_date']=df['registration_date'].bfill()
```

```python
df.info()
```

```python
df['phone']=df['phone'].astype(dtype='object')
df['email']=df['email'].astype(dtype='object')
df['registration_date']=pd.to_datetime(df['registration_date'])
```

```python
df['state']=df['state'].replace('North Carolina', 'NC')
df['state']=df['state'].replace('Texas', 'TX')
df['state']=df['state'].replace('California', 'CA')
df['state']=df['state'].replace('Washington', 'WA')
df['state']=df['state'].replace('Oregon', 'OR')
df['state']=df['state'].replace('Maryland', 'MA')
```
