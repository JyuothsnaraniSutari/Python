```python
shopping_list = ["apples", "bread", "banana", "milk", "eggs", "cheese"]
print(f"The original shopping list is: {shopping_list}")
```

```python
print(f"The third item in my shopping list is: {shopping_list[2]}")
```

```python
shopping_list[0] = "carrots"
print(f"My updated shopping list is: {shopping_list}")
```

```python
shopping_list.append("chicken")
print(f"My updated shopping list is: {shopping_list}")
```

```python
grocery_prices ={
    "apples": 1.29,
    "bread": 2.49,
    "banana": 0.79,
    "milk": 3.19,
    "eggs": 2.99,
    "cheese": 4.49
}
print(f"The grocery prices are: {grocery_prices}")
```

```python
print(f"The price of milk is ${grocery_prices['milk']}")
```

```python
grocery_prices["chicken"] = 2.99
print(f"The updated prices: ${grocery_prices}")
```

```python
total_cost = sum(grocery_prices.values())
print(f"The total cost of all items is: ${total_cost:.2f}")
```
